# Decision Tree

```python
# DECISION TREE
from sklearn.model_selection import cross_val_score
from sklearn.tree import DecisionTreeClassifier
tree_clf = DecisionTreeClassifier()
tree_clf.fit(X_train, y_train)
y_pred_DT = tree_clf.predict(X_test)
```

```python
# Librerias Evaluación
from sklearn import metrics
from sklearn.metrics import roc_curve, auc
from sklearn.metrics import classification_report, confusion_matrix
# Function from scklearn
import itertools
def plot_confusion_matrix(cm, classes, normalize=False,
                          title='Confusion matrix',
                          cmap=plt.cm.Blues):
    """"
    This function prints and plots the confusion matrix.
    """
    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=45)
    plt.yticks(tick_marks, classes)
    
    if normalize:
        cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
        print("Normalized confusion matrix")
    else:
        print('Confusion matrix, without normalization')
        
    print(cm)
    
    thresh = cm.max() /2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j,i, cm[i,j],
                horizontalalignment="center",
                color="white" if cm[i,j] > thresh else "black")
        
    plt.tight_layout()
    plt.ylabel('True label')
    plt.xlabel('Predicted label')
```

```python
# DECISION TREE

# Classification Report
print('DECISION TREE')
print(classification_report(y_test, y_pred_DT))

# Confusion Matrix
cm_dectree = confusion_matrix(y_test, y_pred_DT)
cm_plot_labels = ['No', 'Yes']
plot_confusion_matrix(cm=cm_dectree, classes=cm_plot_labels, title='Decision Tree Confusion Matrix')

# Espacio ROC
dectree_fpr, dectree_tpr, threshold = roc_curve(y_test, y_pred_DT)
auc_dectree = auc(dectree_fpr, dectree_tpr) 
plt.figure(figsize=(5,5), dpi=100)
plt.plot(dectree_fpr, dectree_tpr, linestyle='-', label='Decision Tree(auc = %0.3f)'%auc_dectree)
plt.plot([0, 1], [0, 1], 'k--')
plt.title("ROC Decision Tree")
plt.xlabel('False Positive Rate -->')
plt.ylabel('True Positive Rate -->')
plt.legend()
plt.show()
```

```python
## OPTIMIZACIÓN
from sklearn.model_selection import GridSearchCV

# DECISION TREE
# Round 1
param_grid = {
    'criterion':['gini', 'entropy'],
    'splitter':['best', 'random'],
    'max_depth':[3, 4, 5],
    'random_state':[0, 1, 2],
    'class_weight':[{0: w} for w in [1, 4, 8]],
    'min_samples_split': [2, 6, 10],
    'min_samples_leaf': [1, 2, 4]
}

optimal_params = GridSearchCV(
    estimator=DecisionTreeClassifier(),
    param_grid=param_grid,
    scoring='roc_auc',
    verbose=0,
    n_jobs=10,
    cv=3
)

optimal_params.fit(X_train, y_train)
print(optimal_params.best_params_)
```

```python
# Round 2
param_grid = {
    'criterion':['entropy'],
    'splitter':['random'],
    'max_depth':[4],
    'random_state':[2],
    'class_weight':[{0: w} for w in [7, 8, 9]],
    'min_samples_split': [10, 11, 12],
    'min_samples_leaf': [4, 5, 6]
}

optimal_params = GridSearchCV(
    estimator=DecisionTreeClassifier(),
    param_grid=param_grid,
    scoring='roc_auc',
    verbose=0,
    n_jobs=10,
    cv=3
)

optimal_params.fit(X_train, y_train)
optimal_params = optimal_params.best_params_
print(optimal_params)
```

```python
del optimal_params["class_weight"]
```

```python
tree_clf = DecisionTreeClassifier(class_weight={1:8}, **optimal_params)

tree_clf.fit(X_train, y_train)

y_pred_DT = tree_clf.predict(X_test)
```

```python
# Classification Report
print('DECISION TREE')
print(classification_report(y_test, y_pred_DT))

# Confusion Matrix
cm_dectree = confusion_matrix(y_test, y_pred_DT)
cm_plot_labels = ['No', 'Yes']
plot_confusion_matrix(cm=cm_dectree, classes=cm_plot_labels, title='Decision Tree Confusion Matrix')

# Espacio ROC
dectree_fpr, dectree_tpr, threshold = roc_curve(y_test, y_pred_DT)
auc_dectree = auc(dectree_fpr, dectree_tpr) 
plt.figure(figsize=(5,5), dpi=100)
plt.plot(dectree_fpr, dectree_tpr, linestyle='-', label='Decision Tree(auc = %0.3f)'%auc_dectree)
plt.plot([0, 1], [0, 1], 'k--')
plt.title("ROC Decision Tree")
plt.xlabel('False Positive Rate -->')
plt.ylabel('True Positive Rate -->')
plt.legend()
plt.show()
```
